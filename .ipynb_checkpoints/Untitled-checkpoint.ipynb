```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import statsmodels.api as sm
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.stattools import acf, pacf
import statsmodels.tsa.stattools as ts

plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['text.color'] = 'G'


transData = pd.read_csv('transactionData.csv', sep=',')
transData['Month'] = pd.to_datetime(transData['Month'], format = '%d/%m/%Y')
transCat = transData['Category'].unique() #list of all categories in transData
numCat = len(transCat)
```

```python
#loading and conveting time series data by setting index as time
transData['Month'] = pd.to_datetime(transData['Month'])
transData.index = transData.Month
transData = transData.drop('Month', axis =1)
transData.head()
```

```python
savingsData = transData['Amount']
savingsData = pd.DataFrame(savingsData.groupby(['Month']).sum())

#visualizing timeseries data
plt.plot(savingsData)
plt.xlabel('time')
plt.ylabel('Amount')
plt.plot()
    
#determine param for ARIMA model
amt = savingsData['Amount']

#acf
acf_1 = acf(amt)[1:20]
test_df = pd.DataFrame([acf_1]).T
test_df.columns = ['Autocorrelation']
test_df.index+=20
test_df.plot(kind='bar')
plt.show()

#pacf
pacf_1 = pacf(amt)[1:20]
test_df = pd.DataFrame([pacf_1]).T
test_df.columns = ['Partial Autocorrelation']
test_df.index+=20
test_df.plot(kind='bar')
plt.show()
```

```python
#divide into train and validation set
train = savingsData[:int(0.8*(len(savingsData)))]
test = savingsData[int(0.8*(len(savingsData))):]

print(train.shape)
print(test.shape)
```

```python
#training model
from pmdarima import auto_arima
model = auto_arima(train, trace=True,start_p=0, start_q=0, start_P=0, start_Q=0,
                  max_p=10, max_q=10, max_P=10, max_Q=10, seasonal=True,
                  stepwise=False, suppress_warnings=True, D=1, max_D=10,
                  error_action='ignore',approximation = False)
#fitting model
model.fit(train)
```
